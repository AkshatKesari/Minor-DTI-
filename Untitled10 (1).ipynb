```python
def read_file_training_dataset_bioassay(path):
	# a line in the file is SMILES score, the first line is the target sequence
	try:
		file = open(path, "r")
	except:
		print('Path Not Found, please double check!')
	target = file.readline()
	if target[-1:] == '\n':
		target = target[:-1]
	X_drug = []
	y = []
	for aline in file:
		values = aline.split()
		X_drug.append(values[0])
		y.append(float(values[1]))
	file.close()
	return np.array(X_drug), target, np.array(y)
```

```python
import torch
from rdkit import Chem
import pandas as pd

# Assuming you already have your DataFrame `df` with the 'Mol' column

# Define the constants and functions (as provided)
ELEM_LIST = ['C', 'N', 'O', 'S', 'F', 'Si', 'P', 'Cl', 'Br', 'Mg', 'Na', 'Ca', 'Fe', 'Al', 'I', 'B', 'K', 'Se', 'Zn', 'H', 'Cu', 'Mn', 'unknown']
ATOM_FDIM = len(ELEM_LIST) + 6 + 5 + 4 + 1
BOND_FDIM = 5 + 6
MAX_NB = 6

def onek_encoding_unk(x, allowable_set):
    if x not in allowable_set:
        x = allowable_set[-1]
    return list(map(lambda s: x == s, allowable_set))

def atom_features(atom):
    return torch.Tensor(onek_encoding_unk(atom.GetSymbol(), ELEM_LIST)
            + onek_encoding_unk(atom.GetDegree(), [0, 1, 2, 3, 4, 5])
            + onek_encoding_unk(atom.GetFormalCharge(), [-1, -2, 1, 2, 0])
            + onek_encoding_unk(int(atom.GetChiralTag()), [0, 1, 2, 3])
            + [atom.GetIsAromatic()])

def bond_features(bond):
    bt = bond.GetBondType()
    stereo = int(bond.GetStereo())
    fbond = [bt == Chem.rdchem.BondType.SINGLE, bt == Chem.rdchem.BondType.DOUBLE,
             bt == Chem.rdchem.BondType.TRIPLE, bt == Chem.rdchem.BondType.AROMATIC,
             bond.IsInRing()]
    fstereo = onek_encoding_unk(stereo, [0, 1, 2, 3, 4, 5])
    return torch.Tensor(fbond + fstereo)

# Function to compute features for a molecule
def compute_features(mol):
    atom_feats = []
    bond_feats = []

    for atom in mol.GetAtoms():
        atom_feats.append(atom_features(atom))

    for bond in mol.GetBonds():
        bond_feats.append(bond_features(bond))

    return atom_feats, bond_feats

# Apply feature computation to each molecule in the DataFrame
df['atom_features'] = df['Mol'].apply(lambda mol: compute_features(mol)[0] if mol else None)
df['bond_features'] = df['Mol'].apply(lambda mol: compute_features(mol)[1] if mol else None)

# Display the updated DataFrame
df.head()
```

```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import rdmolops

# Function to convert SMILES to molecular graph (adjacency matrix)
def mol_to_adjacency(smiles):
    """Converts a SMILES string into an adjacency matrix using RDKit."""
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return None  # In case the SMILES string is invalid
    adj_matrix = rdmolops.GetAdjacencyMatrix(mol)
    return adj_matrix

# Assuming you have a DataFrame `df` with a SMILES column
df['Adjacency_Matrix'] = df['SMILES'].apply(mol_to_adjacency)

# View the DataFrame
df.head()
```
